```python
## Revert Time of Conversation and Average Revert Time of Conversations

# stud_df['Date_time'] = datetime.datetime.combine(stud_df.loc[0,'Date'], stud_df.loc[0,'Time'])
# ment_df['Date_time'] = datetime.datetime.combine(ment_df.loc[0,'Date'], ment_df.loc[0,'Time'])

for i in range(stud_df.shape[0]):
    stud_df.loc[i,'Date_time'] = datetime.datetime.combine(stud_df.loc[i,'Date'].date(), stud_df.loc[i,'Time'])

for i in range(ment_df.shape[0]):
    ment_df.loc[i,'Date_time'] = datetime.datetime.combine(ment_df.loc[i,'Date'].date(), ment_df.loc[i,'Time'])
```

```python
stud_rid = stud_df['rid'].unique()
ment_rid = ment_df['rid'].unique()
lst = []
for i in stud_rid:
    df1 = stud_df.groupby('rid').get_group(i)
    if i in ment_rid:
        df2 = ment_df.groupby('rid').get_group(i)
    else:
        continue
    df3 = pd.concat([df1,df2],axis=0)
    df4 = df3.sort_values(by=['Date_time']).reset_index().drop(columns=['index','rType'])

    def response_time(df):
        r = []
        flag=0
        for i in range(0,df.shape[0]):
            if df.loc[i,'u.mentor']==False:
                flag=1
                start_index=i
                break
        if flag==1:
            for i in range(start_index,df.shape[0]):
                if df.loc[i,'u.mentor']==True and df.loc[i-1,'u.mentor']!=True:
                    r.append((df.loc[i,'Date_time']-df.loc[i-1,'Date_time']).total_seconds())
        if len(r)==0:
            return 0
        else:
            response = mean(r)
            return response
    t = response_time(df4)
    lst.append(t)

fig, ax=plt.subplots(figsize=(16,5))
ax.plot(range(len(lst)),lst)

plt.title('Response time per conversation')
plt.xlabel('Conversation count')
plt.ylabel('Time (1 * 10^6 sec)')
```
