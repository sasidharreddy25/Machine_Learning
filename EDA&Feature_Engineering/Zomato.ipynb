```python
df1=pd.read_excel('Country-Code.xlsx')
```

```python
df1
```

```python
df2=pd.merge(df, df1, on='Country Code', how='left')
```

```python
coutry_names=df2.Country.value_counts().index
```

```python
country_va=df2.Country.value_counts().values
```

```python
#pie chart
plt.pie(x=country_va[:3], labels=coutry_names[:3], autopct='%1.2f%%')
```

observation: Most transactions are from India followed by us, uk

```python
ratings=df2.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'rating_count'})
```

```python
ratings
```

Observation: rating b/w 4.5 to 4.9------>Excellent
4 to 4.5 -------> very good
3.5 to 4 -------> good
3 to 3.5 --------> average
2.5 to 3 --------> poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='rating_count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='rating_count', hue='Rating color', data=ratings, palette=['White','Red','Orange','Yellow','Green','Green'])
```

observations: not rated is high and maximum rating is b/w 2.4 to 3.9

```python
#count plot
sns.countplot(x='Rating color',data=ratings, palette=['blue','Red','Orange','Yellow','Green','Green'])
```
